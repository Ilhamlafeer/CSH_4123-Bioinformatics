```python
seq = 'ATGTCCTTTTTATCG'
complement = ''

for i in seq:
    if i =='A':
        complement += 'T'
    elif i == 'T':
        complement += 'A'
    elif i == 'C':
        complement += 'G'
    else:
        complement += 'C'
print(f"Complement Sequence is: {complement}")
```

```python
# Transcription (DNA to mRNA)
transcription = ''
for i in complement:
    if i =='T':
        transcription += 'U'
    else:
        transcription += i
print(f"mRNA Sequence is: {transcription}")
```

```python
# Codons
def group_letters(transcription):
    groups = []
    for i in range(0, len(transcription), 3):
        groups.append(transcription[i:i+3])
    return groups

groups = group_letters(transcription)
print(groups)
```
