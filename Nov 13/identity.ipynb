```python
# qsn 1
seq1 = 'ATGCATAG'
seq2 = 'ATGAATGG'

def matching(seq1, seq2):
    match = 0
    # Check identity (identity = No of exacat matches/Total positions)*100 
    for i in range(len(seq1)):
        if seq1[i] == seq2[i]:
            match+=1

    total = len(seq1)
    percent_identity = (match/total)*100
    print(f"Total Sequence length: {total}")
    print(f"Matching count: {match}")
    print(f"Percentage: {percent_identity}%")

matching(seq1, seq2)
```

```python
# qsn 2
DNA_seq1 = 'ATGGCTGAACTG'
DNA_seq2 = 'ATGGCCGAATTG'

# Transcription
def transcript(sequence):
    mrna = ''
    for i in sequence:
        if i == 'A':
            mrna += 'U'
        elif i == 'T':
            mrna += 'A'   
        elif i == 'C':
            mrna += 'G'  
        else:
            mrna += 'C'
    return mrna
    
# Codons
def group_letters(mrna):
    groups = []
    for i in range(0, len(mrna), 3):
        groups.append(mrna[i:i+3])
    return groups

# Translation (Protein)
def protein(groups):
    amino = []
    for i in groups:
        if i in ['UUU', 'UUC']:
            amino.append('Phe')
        elif i in ['UUA', 'UUG', 'CUU', 'CUC', 'CUA', 'CUG']:
            amino.append('Leu')
        elif i in ['AUU', 'AUC', 'AUA']:
            amino.append('Ile')
        elif i == 'AUG':
            amino.append('Met')
        elif i in ['GUU', 'GUC', 'GUA', 'GUG']:
            amino.append('Val')
        elif i in ['UCU', 'UCC', 'UCA', 'UCG', 'AGU', 'AGC']:
            amino.append('Ser')
        elif i in ['CCU', 'CCC', 'CCA', 'CCG']:
            amino.append('Pro')
        elif i in ['ACU', 'ACC', 'ACA', 'ACG']:
            amino.append('Thr')
        elif i in ['GCU', 'GCC', 'GCA', 'GCG']:
            amino.append('Ala')
        elif i in ['UAU', 'UAC']:
            amino.append('Tyr')
        elif i in ['UAA', 'UAG', 'UGA']:
            amino.append('Stop')
        elif i in ['CAU', 'CAC']:
            amino.append('His')
        elif i in ['CAA', 'CAG']:
            amino.append('Gln')
        elif i in ['AAU', 'AAC']:
            amino.append('Asn')
        elif i in ['AAA', 'AAG']:
            amino.append('Lys')
        elif i in ['GAU', 'GAC']:
            amino.append('Asp')
        elif i in ['GAA', 'GAG']:
            amino.append('Glu')
        elif i in ['UGU', 'UGC']:
            amino.append('Cys')
        elif i == 'UGG':
            amino.append('Trp')
        elif i in ['CGU', 'CGC', 'CGA', 'CGG', 'AGA', 'AGG']:
            amino.append('Arg')
        elif i in ['GGU', 'GGC', 'GGA', 'GGG']:
            amino.append('Gly')
        else:
            amino.append('??? - ')
    return amino
    
    
transcript_1 = transcript(DNA_seq1)
print(f"Transcription of DNA Sequence1: {transcript_1}")

transcript_2 = transcript(DNA_seq2)
print(f"Transcription of DNA Sequence1: {transcript_2}\n")

codon_1 = group_letters(transcript_1)
print(f"Codons of DNA Sequence1: {codon_1}")
codon_2 = group_letters(transcript_2)
print(f"Codons of DNA Sequence2: {codon_2}\n")

translation_1 = protein(codon_1)
print(f"Amino sequence of DNA Sequence1: {translation_1}")
translation_2 = protein(codon_2)
print(f"Amino sequence of DNA Sequence2: {translation_2}\n")

# Amino acid similarity
```

```python
# Manual DNA mutation and translation

DNA_Seq = "ATGCCGTAGCTAAGGCTT"

# --- Mutations ---
# Substitution: replace base at position 5 with 'A'
sub_seq = DNA_Seq[:5] + "A" + DNA_Seq[6:]

# Insertion: insert 'T' after position 5
ins_seq = DNA_Seq[:5] + "A" + DNA_Seq[5:]

# Deletion: remove base at position 5
del_seq = DNA_Seq[:5] + DNA_Seq[6:]

# --- Codon to Amino Acid Table (simplified) ---
codon_table = {
    "ATA":"I","ATC":"I","ATT":"I","ATG":"M",
    "ACA":"T","ACC":"T","ACG":"T","ACT":"T",
    "AAC":"N","AAT":"N","AAA":"K","AAG":"K",
    "AGC":"S","AGT":"S","AGA":"R","AGG":"R",
    "CTA":"L","CTC":"L","CTG":"L","CTT":"L",
    "CCA":"P","CCC":"P","CCG":"P","CCT":"P",
    "CAC":"H","CAT":"H","CAA":"Q","CAG":"Q",
    "CGA":"R","CGC":"R","CGG":"R","CGT":"R",
    "GTA":"V","GTC":"V","GTG":"V","GTT":"V",
    "GCA":"A","GCC":"A","GCG":"A","GCT":"A",
    "GAC":"D","GAT":"D","GAA":"E","GAG":"E",
    "GGA":"G","GGC":"G","GGG":"G","GGT":"G",
    "TCA":"S","TCC":"S","TCG":"S","TCT":"S",
    "TTC":"F","TTT":"F","TTA":"L","TTG":"L",
    "TAC":"Y","TAT":"Y","TAA":"STOP","TAG":"STOP",
    "TGC":"C","TGT":"C","TGA":"STOP","TGG":"W",
}

# --- Translation Function ---
def translate_dna(seq):
    protein = ""
    for i in range(0, len(seq), 3):
        codon = seq[i:i+3]
        if len(codon) < 3:
            break
        aa = codon_table.get(codon, "")
        if aa == "STOP":
            break
        protein += aa
    return protein

# --- Translate all sequences ---
original_protein = translate_dna(DNA_Seq)
sub_protein = translate_dna(sub_seq)
ins_protein = translate_dna(ins_seq)
del_protein = translate_dna(del_seq)

# --- Identity Percentage ---
def identity_percentage(seq1, seq2):
    matches = sum(1 for a, b in zip(seq1, seq2) if a == b)
    return matches / max(len(seq1), len(seq2)) * 100

# --- Results ---
print("Proteins:")
print("Original:", original_protein, "Length:", len(original_protein))
print("Substitution:", sub_protein, "Length:", len(sub_protein))
print("Insertion:", ins_protein, "Length:", len(ins_protein))
print("Deletion:", del_protein, "Length:", len(del_protein))

print("\nDNA Identity Percentages:")
print("Original vs Substitution:", identity_percentage(DNA_Seq, sub_seq))
print("Original vs Insertion:", identity_percentage(DNA_Seq, ins_seq))
print("Original vs Deletion:", identity_percentage(DNA_Seq, del_seq))
```
